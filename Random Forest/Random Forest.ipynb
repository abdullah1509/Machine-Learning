```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
df= pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
print(iris)
```

```python
df['species']= pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
#Creating Test and Train data
df['is_train']= np.random.uniform(0,1, len(df))<= .75
df.head()
```

```python
#Creating dataframes with test row and training row
train, test= df[df['is_train']== True], df[df['is_train']== False]

#Show number of observations for test and training dataframes
print('No of observatio in Training data: ', len(train))
print('No of observatio in Test data: ', len(test))
```

```python
#Create list of feature column's name
features= df.columns[:4]
features
```

```python
#Converting each species in digits
y= pd.factorize(train['species'])[0]
y
```

```python
#Creating Random Forest Classifier
clf= RandomForestClassifier(n_jobs=2, random_state=0)

#Training Classifier
clf.fit(train[features], y)
```

```python
#Applying trained classifier to the test
clf.predict(test[features])
```

```python
features
```

```python
test[features]
```

```python
#Viewing predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
#Mapping names for the plants for each predicted plant class
preds= iris.target_names[clf.predict(test[features])]
```

```python
#View the predicted species
preds
```

```python
#Viewing actual species
test['species']
```

```python
#Confusion Matrix
pd.crosstab(test['species'], preds, rownames=['Actual Species'], colnames=['Predicted Species'])
```

```python
#Testing predictions
preds= iris.target_names[clf.predict([[5.0, 3.6, 1.4, 2.0], [2.0, 1.6, 0.4, 0.0]])]
preds
```
