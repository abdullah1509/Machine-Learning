```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
from sklearn.datasets import fetch_20newsgroups
data= fetch_20newsgroups()
data.target_names
```

```python
#define all categories
categories=['alt.atheism',
 'comp.graphics', 'comp.os.ms-windows.misc', 'comp.sys.ibm.pc.hardware', 'comp.sys.mac.hardware', 'comp.windows.x',
 'misc.forsale', 'rec.autos', 'rec.motorcycles', 'rec.sport.baseball', 'rec.sport.hockey', 'sci.crypt', 'sci.electronics',
 'sci.med', 'sci.space', 'soc.religion.christian', 'talk.politics.guns', 'talk.politics.mideast', 'talk.politics.misc',
 'talk.religion.misc']
```

```python
#Training data
train= fetch_20newsgroups(subset='train', categories=categories)
```

```python
#Testing data
test= fetch_20newsgroups(subset='test', categories=categories)
```

```python
print(train.data[5])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
#create a model
model= make_pipeline(TfidfVectorizer(), MultinomialNB())

#Training model
model.fit(train.data, train.target)

#creating labels
labels= model.predict(test.data)
```

```python
#creating confusion matrix
from sklearn.metrics import confusion_matrix
mat= confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, 
           xticklabels=train.target_names, yticklabels=train.target_names, linewidths=.5)

plt.xlabel('true label')
plt.ylabel('predicted label')
```

```python
#Predicting category
def predict_category(s, train=train, model=model):
    pred= model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category('University')
```
