### Using Plotly 

```python
import plotly.graph_objects as go
import pandas as pd

data = pd.read_csv('diabetes.csv')
data.head()
```

```python
fig = go.Figure(go.Waterfall(
    name = "20", orientation = "v",
    measure = ["total", "relative", "relative", "relative", "relative", "total"],
    x = ["Pregnancies", "Age", "Glucose", "BloodPressure", "BMI", " Insulin"],
    #textposition = "outside",
    y = [data['Pregnancies'], data['Age'], data['Glucose'], data['BloodPressure'], data['BMI'], data['Insulin']],
    connector = {"line":{"color":"rgb(63, 63, 63)"}},
))

fig.update_layout(
       # title = "Profit and loss statement 2018",
        showlegend = True
)

fig.show()
```

### Using Bokeh

```python
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
output_notebook()


p = figure(y_range='Pregnancies', x_range='BMI', plot_width=400, plot_height=550,
           title="Time Spreads for Sprint Medalists (by Year)")
p.hbar(y='Pregnancies', left='Age', right='Insulin', height=0.4, source=data)


p.ygrid.grid_line_color = None
p.xaxis.axis_label = "Time (seconds)"
p.outline_line_color = None

show(p)
```
