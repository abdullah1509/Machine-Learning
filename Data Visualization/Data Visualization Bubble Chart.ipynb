### Using Plotly 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('winequality-red.csv')
data.head()
```

```python
import plotly.express as px

fig = px.scatter( x=data['fixed acidity'], y=data['alcohol'],size=data['density'], 
                 color=data['residual sugar'],hover_name=data['pH'],  size_max=60)
fig.show()
```
