### Using Matplotlib

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")

x = [15,20,30]
y = [10,25,35]

x1 = [29,25,15]
y1 = [10,20,30]

plt.title("Info")
plt.xlabel("X-axix")
plt.ylabel("Y-axix")
plt.plot(x,y, 'red', label='red line')
plt.plot(x1,y1, 'blue', label='blue line')

plt.legend()
plt.grid(True, color='k')
plt.show()
```

```python
import pandas as pd

data=pd.read_csv('50_Startups.csv')
data.head()
```

```python
plt.figure(figsize=(16,10))

plt.plot(data['State'], data['Marketing Spend'], color='red', marker='x', label='Expenditure')
plt.plot(data['State'], data['Profit'], color='green', marker='o', label='Profit')
plt.legend()
plt.title("Expenditure Vs Profit")
plt.xlabel('State')
plt.ylabel('Amount')

plt.show()
```

### Using Plotly

```python
import plotly.graph_objects as go
import plotly.express as px


px.line(data, x='State', y='Marketing Spend', labels={'x':'State', 'y':'Expenditure'})
px.line(data, x='State', y=['Marketing Spend', 'Profit'], labels={'x':'State', 'y':'Profit'}, title="Exp Vs Profit")
```

### Using Seaborn

```python
import numpy as np
import seaborn as sns

sns.relplot(x='State', y='Profit', data=data, kind='line')
```

### Using Bokeh

```python
from bokeh.io import output_notebook, show, reset_output
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
output_notebook()


line_plot = figure(plot_width=500, plot_height=325, title='Exp Vs Profit', x_axis_label='State', y_axis_label='Profit')

line_plot.line(data['State'], data['Profit'], legend='Profit', line_width=2, color='red')
line_plot.line(x=['State'], y=['Marketing Spend'],legend='Expenditure', line_width=2, color='blue')
line_plot.line(x=['State'], y=['R&D Spend'],legend='R&D', line_width=2, color='green')

line_plot.add_tools(HoverTool())
show(line_plot)
#There must be some error in data
```
