### Using Plotly

```python
import pandas as pd
data= pd.read_csv('API_ILO_country_YU.csv')
data.head()
```

```python
#Simple Bubble Chart
import plotly.graph_objects as go
fig = go.Figure(data=[go.Scatter( x=data['2010'], y=data['2014'], mode='markers')])
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()

fig = px.scatter(data, x="2010", y="2012", size="2014", 
                 color="Country Code", hover_name="Country Name", log_x=True, size_max=60)
fig.show()
```

### Using Seaborn

```python
import seaborn as sns
sns.set(style="white")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="2010", y="2014", hue="Country Code", size="2014",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=data)
```

### Using Altair

```python
import altair as alt

alt.Chart(data).mark_point().encode( x='2010', y='2014', size='Country Code')
```
