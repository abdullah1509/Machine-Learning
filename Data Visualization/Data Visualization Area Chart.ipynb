### Using Matplotlib 

```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('mtcars.csv')
data.head()
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,10), dpi=80)
plt.stackplot(data['carb'],data['cyl'], data['mpg'], data['hp'], colors=['r','g','b','c'])

X = data['car']
Y= data['disp']
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Stack Plot")
plt.show()
```

### Using Seaborn

```python
import numpy as np
import seaborn as sns
import pandas as pd

#for col in data:
sns.kdeplot(data['hp'], shade=True)
sns.kdeplot(data['mpg'], shade=True)
sns.kdeplot(data['qsec'], shade=True)
```

### Using Bokeh

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook, show
output_notebook()

p=figure(plot_width=600, plot_height=400)
p.varea_stack(['hp', 'mpg', 'carb'], x='disp', color=['pink', 'skyblue', 'DarkViolet'], source=data)
show(p)
```

### Using Altair

```python
import altair as alt

alt.Chart(data).mark_area().encode(x='disp', y='hp')
```
